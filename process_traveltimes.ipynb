The code below reads the zipped file into json.

```python
## The code below reads the zipped file and produces a (geo?)json file.
import json
from zipfile import ZipFile
import zipfile_deflate64
import pandas as pd
import geopandas as gpd
import rioxarray as rio
from geocube.api.core import make_geocube
```

```python
origin_points = {}
```

```python
with ZipFile('travel_times_walking.zip') as zfile:
    traveltimes_ifile = list(zfile.namelist())[0]
    with zfile.open(traveltimes_ifile) as ifile:
        linecount = 0
        print(ifile.readline())
        keep_going = True
        while keep_going:
            linecount += 1
            line = ifile.readline()
            keep_going = line is not None
            if keep_going:
                items = str(line).split(',')
                traveltime_str = items[2].strip()
                if traveltime_str[:-5] != "":
                    idnum = int(items[0][2:])
                    traveltime = float(traveltime_str[:-5])
                    if not idnum in list(origin_points.keys()):
                        origin_points[idnum] = traveltime
                    else:
                        origin_points[idnum] = min(origin_points[idnum], traveltime)
            if linecount % 100000 == 0:
                print(linecount)
```

```python
import json
with open('traveltimes.json', 'w') as ofile:
    ofile.write(json.dumps(origin_points))
```

```python
traveltimes_pd = pd.DataFrame({'idnum': origin_points.keys(), 'walktime': origin_points.values()})
```

```python
traveltimes_pd
```

```python
worldpopcells_old = gpd.read_file('worldpop_cells.geojson')
```

```python
worldpopcells = worldpopcells_old.copy(deep=True)
```

```python
worldpopcells['walk_time'] = -9999
```

```python
worldpopcells
```

```python
for idx in range(len(traveltimes_pd)):
    from_row = traveltimes_pd.iloc[idx]
    worldpopcells.loc[worldpopcells['pointid']==from_row['idnum'], ['walk_time']] = from_row['walktime']
```

```python
walktimes_gpd = worldpopcells.loc[worldpopcells['walk_time'] != -9999]
```

```python
walktimes_gpd
```

```python
out_ras = make_geocube(vector_data = walktimes_gpd, measurements=["walk_time"], resolution=(0.000835,0.000835))
```

```python
out_ras["walk_time"].rio.to_raster("guadalajara_walktime.tif")
```
